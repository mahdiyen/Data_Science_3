```python
#perform log transformation on skewed_columns
# the indexes are the columns of the data, and the second column is the amount of skewness
skew_columns = (data.skew().sort_values(ascending = False)) # make the first column as the index
skew_columns = skew_columns[skew_columns > 0.75] # keep the ones that are more than 0.75

# calculate the log transformation
# this replace each column data with the log transform
for col in skew_columns.index.tolist():
    data[col] = np.log1p(data[col])
data.shape
```

## Pipeline: 
Define the pipeline with preprocessing and clustering:
When I want to use t-SNE and clustering together in pipeline, I get an error. It seems that the t-SNE algorithm does not implement the transform method required by the pipeline in this way. So, I have two approaches:

1- use pipeline for normalization, PCA, and t_SNE , Then clustering

2- use pipeline for clustering without t-SNE

### 1- Pipeline for normalization, PCA, and t_SNE , Then clustering 

```python
preprocessor = Pipeline([
    ('scaler', StandardScaler()),   # Standardize features
    ('pca', PCA(n_components=2)),   # Apply PCA for dimensionality reduction
    ('tsne', TSNE(n_components=2)), # Apply t-SNE for dimensionality reduction
])
```

```python
reduced_data = preprocessor.fit_transform(data)
```

```python
# K-means Clustering
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters)
cluster_labels = kmeans.fit_predict(reduced_data)
```

```python
# plotting
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=cluster_labels)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Dimension Reduction with Clustering')
plt.show()
```

### 2- Pipeline without t-SNE

```python
# pipeline definition for normalization, PCA and clustering
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)), 
    ('kmeans', KMeans(n_clusters=5))
])
```

```python
reduced_data1 = pipeline.fit_transform(data)
cluster_labels1 = kmeans.fit_predict(reduced_data1)
```

```python
# plotting
plt.scatter(reduced_data1[:, 0], reduced_data1[:, 1], c=cluster_labels1)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Dimension Reduction with Clustering')
plt.show()
```

## Evaluation:
#### Silhouette Score: 
The silhouette score measures the compactness and separation of clusters. A higher silhouette score indicates better-defined and well-separated clusters.

```python
from sklearn.metrics import silhouette_score

silhouette_score = silhouette_score(data, cluster_labels)
print("Silhouette Score (Pipeline 1):", silhouette_score)
```

# gridsearchCV

#### gridsearchCV on the second approach

```python
hyperparams = {
    "kmeans__n_clusters": [2, 3, 4, 5, 6],
    "kmeans__random_state": [42],
    "kmeans__init": ['k-means++', 'random']
}
```

```python
# pipeline definition for normalization, PCA and clustering
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)), 
    ('kmeans', KMeans())
])
```

```python
# Instantiate the GridSearchCV object with the Pipeline object and parameter grid
grid_search = GridSearchCV(estimator=pipeline, param_grid=hyperparams, cv=10, n_jobs=-1)

# Fit the GridSearchCV object to the data
grid_search.fit(data)

# Access the best combination of hyperparameters and the corresponding evaluation metric score
print(grid_search.best_params_)
print(grid_search.best_score_)
```

## Explanation and conclusion:
GridSearchCV is a class in scikit-learn that performs an exhaustive search over a specified parameter grid to find the best combination of hyperparameters for a given estimator.

I used this class to find the best combination Kmeans parameters. The results shows that:

The best k-mean clustering is with 'kmeans__init': 'k-means++', 'kmeans__n_clusters': 6, 'kmeans__random_state': 42.
